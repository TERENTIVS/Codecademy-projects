TE: How does the average value of questions containing `England` compare to the overall average question value?

TE: At first glance popular categories seem to take up a large fraction of our dataset; though how much really? We can look at the fraction of the dataset contained within categories with more than a certain number of questions: 

```python
print((jep_topcats[jep_topcats["Count"] > 50].Count.sum()) /
      jep_topcats.Count.sum())
print((jep_topcats[jep_topcats["Count"] > 100].Count.sum()) /
      jep_topcats.Count.sum())
print((jep_topcats[jep_topcats["Count"] > 250].Count.sum()) /
      jep_topcats.Count.sum())
```

TE: Wow! Over more than three decades, the show has maintained a huge diversity of questions.
According to the above result, only about 20% of questions have come from categories that have generated more than 50 questions; less than 15% have come from categories that have generated more than 100 questions; and only about 5% of questions have come from categories with more than 250 questions.

TE: Another question we might be interested in answering is how the top categories changed through history, from the 1980s, through the 1990s and to the end of the dataset. Such changes may be a good reflection of changes in the topics that hold public attention.

```python
# Let's first see when exactly our dataset begins:

print(jep_bydate["Air Date"].min())
```

```python
# Filtering for the 1980s:

jep_80s = jep_bydate[(jep_bydate["Air Date"] >= "1984-09-10") & (jep_bydate["Air Date"] < "1990-01-01")].\
reset_index()

jep80s_topcats = jep_80s["Category"].value_counts().reset_index()
print(jep80s_topcats.head(10))
```

```python
# Now for the 1990s:

jep_90s = jep_bydate[(jep_bydate["Air Date"] > "1989-31-12") & (jep_bydate["Air Date"] < "2000-01-01")].\
reset_index()

jep90s_topcats = jep_90s["Category"].value_counts().reset_index()
print(jep90s_topcats.head(10))
```

```python
# Into the first decade of the new millenium and beyond:

jep_millenium = jep_bydate[(jep_bydate["Air Date"] >= "2000-01-01")].\
reset_index()

jep_millenium_topcats = jep_millenium["Category"].value_counts().reset_index()
print(jep_millenium_topcats.head(10))
```
